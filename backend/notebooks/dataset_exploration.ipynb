```python
import pandas as pd
from pathlib import Path

cwd = Path.cwd()
DATA_DIR = cwd.parent / "data" / "processed"

reviews = pd.read_csv(DATA_DIR / "reviews_processed.csv", low_memory=False)
products = pd.read_csv(DATA_DIR / "products_processed.csv", low_memory=False)

# normalize ids
reviews_ids = set(reviews["product_id"].astype(str).str.strip().dropna().unique())
product_ids = set(products["product_id"].astype(str).str.strip().dropna().unique())

# product_ids present in products file but missing in reviews file
zero_review_ids = sorted(product_ids - reviews_ids)

print("Total products:", len(product_ids))
print("Products with >=1 review row:", len(product_ids & reviews_ids))
print("Products with ZERO review rows:", len(zero_review_ids))

# show the ids (prints all; if too many, slice like [:200])
print(zero_review_ids)
```

```python
import pandas as pd

cwd = Path.cwd()
DATA_DIR = cwd.parent / "data" / "processed"

reviews = pd.read_csv(DATA_DIR / "reviews_processed.csv", low_memory=False)
sample = reviews[
    reviews["product_id"] == "P474806"
]

products = pd.read_csv(DATA_DIR / "products_processed.csv")

product_cols = [
    "product_id",
    "product_name",
    "brand_name",
    "price_usd",
    "primary_category",
    "secondary_category",
    "tertiary_category",
]

lookup = products[product_cols].drop_duplicates(
    subset=["product_id"]
)

reviews = reviews.drop(
    columns=[
        c for c in product_cols
        if c != "product_id" and c in reviews.columns
    ],
    errors="ignore"
)

reviews = reviews.merge(
    lookup,
    on="product_id",
    how="left"
)

reviews["product_name_lower"] = (
    reviews["product_name"]
    .fillna("")
    .str.lower()
    .str.strip()
)

reviews.to_csv(
    DATA_DIR / "reviews_processed.csv",
    index=False
)

print("Done")
```

```python
import requests

def get_sephora_image(product_id: str) -> str | None:
    """product_id like '1720143' (numeric Sephora SKU)"""
    clean_id = str(product_id).lstrip("Pp")
    url = f"https://www.sephora.com/productimages/sku/s{clean_id}-main-zoom.jpg"
    try:
        r = requests.head(url, timeout=5, headers={"User-Agent": "Mozilla/5.0"})
        if r.status_code == 200:
            return url
    except Exception:
        pass
    return None
```
